```python
#Import libraries
import numpy as np 
import pandas as pd 

#Read clean data 
sales= pd.read_csv('Clean_sales.csv')
forecast= pd.read_csv('Clean_forecast.csv')
```

# Data model

```python
data_model=sales.merge(forecast,how= "left", on=["CountryRegion","Brand","Year"])
#check 
data_model
```

```python
#Validation if the releationship is correct or no 
#and i will do it by get the Groupby of data_model when year =2009 same in forecast data after that i will compere two columns data_model.Forecast with forecast.Forecast
valid_model = data_model[data_model["Year"]==2009].groupby(
                                            ["CountryRegion",
                                             "Brand",
                                             "Year"]
                                            ).agg({"Forecast":"mean"}).rename(columns={
                                                                                        "Price":"Price"
                                                                                        }).reset_index()
valid_model
```

```python
valid_model["Forecast_true"] = forecast.Forecast
valid_model
```

```python
def valiation(x,y): #This function to return if Forecast and Forecast_true same or not 
    if x == y :
        return True
    else:
        return False
valid_model["validation"] = valid_model.apply(lambda x: valiation(x["Forecast"], x["Forecast_true"]), axis=1)
#show there is Fasle meaning there is value not same or no
valid_model["validation"].unique()
```

**Note**

After this valiation we find the data_model data is correct relationship to i can use it as date model and this data connect it with both Sales and Forecast.
